1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter.

```python
class Counter:
    def __init__(self, max_value):
        self.max_value = max_value
        self.count = 0

    def increment(self):
        if self.count >= self.max_value:
            print("Error!!! The counter is already at its maximum limit.")
        else:
            self.count += 1

    def reset(self):
        self.count = 0


# Testing Counter class
counter1 = Counter(3)

counter1.increment()
print("Value after first increment:", counter1.count)

counter1.increment()
counter1.increment()
print("Value after three increments:", counter1.count)

print("Attempting one more increment...")
counter1.increment()

counter1.reset()
print("Value after resetting:", counter1.count)
```

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.grid = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.grid[row][col] = char

    def get_pixel(self, row, col):
        return self.grid[row][col]

    def clear(self):
        self.grid = [[' '] * self.width for _ in range(self.height)]

    def display(self):
        print("\n".join(["".join(r) for r in self.grid]))

class Shape:
    def get_x(self): raise NotImplementedError()
    def get_y(self): raise NotImplementedError()
    def get_area(self): raise NotImplementedError()
    def get_perimeter(self): raise NotImplementedError()
    def get_perimeter_points(self): raise NotImplementedError()
    def is_inside(self, x, y): raise NotImplementedError()
    def overlaps(self, other):
        for px, py in self.get_perimeter_points():
            if other.is_inside(px, py):
                return True
        for px, py in other.get_perimeter_points():
            if self.is_inside(px, py):
                return True
        return False
    def paint(self, canvas, char='*'): raise NotImplementedError()

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_area(self): return self.__length * self.__width
    def get_perimeter(self): return 2 * (self.__length + self.__width)

    def is_inside(self, x, y):
        return self.__x <= x <= self.__x + self.__width and self.__y <= y <= self.__y + self.__length

    def paint(self, canvas, char='*'):
        x0, y0 = int(self.__x), int(self.__y)
        w, h = int(self.__width), int(self.__length)
        # top and bottom
        for i in range(w + 1):
            if 0 <= y0 < canvas.height and 0 <= x0 + i < canvas.width:
                canvas.set_pixel(y0, x0 + i, char)
            if 0 <= y0 + h < canvas.height and 0 <= x0 + i < canvas.width:
                canvas.set_pixel(y0 + h, x0 + i, char)
        # sides
        for i in range(h + 1):
            if 0 <= y0 + i < canvas.height and 0 <= x0 < canvas.width:
                canvas.set_pixel(y0 + i, x0, char)
            if 0 <= y0 + i < canvas.height and 0 <= x0 + w < canvas.width:
                canvas.set_pixel(y0 + i, x0 + w, char)

class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_area(self): return math.pi * self.__radius**2
    def get_perimeter(self): return 2 * math.pi * self.__radius

    def is_inside(self, x, y):
        return (x - self.__x)**2 + (y - self.__y)**2 <= self.__radius**2

    def paint(self, canvas, char='*'):
        cx, cy = int(self.__x), int(self.__y)
        r = self.__radius
        total_points = int(2 * math.pi * r * 3)  # denser points
        for i in range(total_points):
            angle = 2 * math.pi * i / total_points
            px = int(round(cx + r * math.cos(angle)))
            py = int(round(cy + r * math.sin(angle)))
            if 0 <= px < canvas.width and 0 <= py < canvas.height:
                canvas.set_pixel(py, px, char)

class Triangle(Shape):
    def __init__(self, a, b, c, x, y):
        self.__a = a
        self.__b = b
        self.__c = c
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_perimeter(self): return self.__a + self.__b + self.__c

    def get_area(self):
        s = self.get_perimeter() / 2
        return math.sqrt(s*(s-self.__a)*(s-self.__b)*(s-self.__c))

    def is_inside(self, x, y):
        x1, y1 = self.__x, self.__y
        x2, y2 = self.__x + self.__a, self.__y
        angle = math.acos((self.__a**2 + self.__b**2 - self.__c**2)/(2*self.__a*self.__b))
        x3 = self.__x + self.__b * math.cos(angle)
        y3 = self.__y + self.__b * math.sin(angle)
        A = self.get_area()
        def tri_area(x1,y1,x2,y2,x3,y3):
            return abs((x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2))/2)
        A1 = tri_area(x, y, x2, y2, x3, y3)
        A2 = tri_area(x1, y1, x, y, x3, y3)
        A3 = tri_area(x1, y1, x2, y2, x, y)
        return abs(A - (A1 + A2 + A3)) < 1e-5

    def paint(self, canvas, char='*'):
        x1, y1 = self.__x, self.__y
        x2, y2 = self.__x + self.__a, self.__y
        angle = math.acos((self.__a**2 + self.__b**2 - self.__c**2)/(2*self.__a*self.__b))
        x3 = self.__x + self.__b * math.cos(angle)
        y3 = self.__y + self.__b * math.sin(angle)

        def draw_line(x_start, y_start, x_end, y_end):
            steps = max(abs(x_end-x_start), abs(y_end-y_start))
            steps = int(math.ceil(steps))
            for i in range(steps + 1):
                t = i/steps if steps>0 else 0
                px = int(round(x_start + t*(x_end-x_start)))
                py = int(round(y_start + t*(y_end-y_start)))
                if 0 <= px < canvas.width and 0 <= py < canvas.height:
                    canvas.set_pixel(py, px, char)

        draw_line(x1, y1, x2, y2)
        draw_line(x2, y2, x3, y3)
        draw_line(x3, y3, x1, y1)

class CompoundShape(Shape):
    def __init__(self, shapes):
        self.shapes = shapes

    def paint(self, canvas, char='*'):
        for s in self.shapes:
            s.paint(canvas, char)

# --- Demo ---
canvas = Canvas(40, 20)
rect = Rectangle(6, 12, 2, 2)
circ = Circle(4, 28, 8)
tri = Triangle(8, 10, 12, 10, 10)
comp = CompoundShape([rect, circ, tri])
comp.paint(canvas, char='#')
canvas.display()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again.

```python
class RasterDrawing:
    def __init__(self):
        self.shapes_dict = {}
        self.shapes_order = []

    def add_shape(self, shape):
        if not hasattr(shape, "name") or shape.name == "":
            shape.name = self._generate_name()
        self.shapes_dict[shape.name] = shape
        self.shapes_order.append(shape.name)

    def _generate_name(self):
        base = "shape"
        name = f"{base}_0"
        i = 1
        while name in self.shapes_dict:
            name = f"{base}_{i}"
            i += 1
        return name

    def paint(self, canvas):
        for name in self.shapes_order:
            self.shapes_dict[name].paint(canvas)

    def update(self, canvas):
        canvas.clear()
        self.paint(canvas)


def rect_set_width(self, new_w):
    self._Rectangle__width = new_w

def rect_set_length(self, new_l):
    self._Rectangle__length = new_l

Rectangle.set_width = rect_set_width
Rectangle.set_length = rect_set_length

# --- Testing RasterDrawing ---
print("--- Initial Drawing ---")
canvas = Canvas(40, 20)
drawing = RasterDrawing()

rect1 = Rectangle(4, 8, 2, 2)
circle1 = Circle(3, 25, 5)

drawing.add_shape(rect1)
drawing.add_shape(circle1)

drawing.paint(canvas)
canvas.display()

print("\n--- Modified Drawing ---")
drawing.shapes_dict["shape_0"].set_width(18)
drawing.shapes_dict["shape_0"].set_length(8)

triangle1 = Triangle(6, 6, 8, 2, 12)
drawing.add_shape(triangle1)

drawing.update(canvas)
canvas.display()
```
