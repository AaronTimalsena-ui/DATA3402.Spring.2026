*Exercise 1:* Write a function that creates an **n×n matrix** (a list of lists) representing the state of a tic‑tac‑toe game.

Use the integers:

- `0` = empty
- `1` = `"X"`
- `2` = `"O"`


````python
# Write your solution here
def create_board_square_matrix(n):
  board = []
  for i in range(n):
    row = []
    for j in range(n):
      row.append(0)
    board.append(row)
  return(board)
````

*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
def check_winner(board):
    n = len(board)
    for row in board:
        if row.count(row[0]) == n and row[0] != 0:
            return row[0]
    for col in range(n):
        first = board[0][col]
        if first != 0:
            win = True
            for row in range(n):
                if board[row][col] != first:
                    win = False
                    break
            if win:
                return first
    first = board[0][0]
    if first != 0:
        win = True
        for i in range(n):
            if board[i][i] != first:
                win = False
                break
        if win:
            return first
    first = board[0][n - 1]
    if first != 0:
        win = True
        for i in range(n):
            if board[i][n - 1 - i] != first:
                win = False
                break
        if win:
            return first
    for row in board:
        if 0 in row:
            return -1
    return 0
````

*Exercise 5:* Write a function that takes a game board, a player number, and `(row, col)` coordinates and places the correct mark (`"X"` or `"O"`) in that location.

Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
# Write your solution here
def place_move(board, player, row, col):
    if board[row][col] == 0:
        board[row][col] = player
        return True
    else:
        return False
````

*Exercise 6:* Modify Exercise 3 to show **row and column labels** so that players can specify locations like `"A2"` or `"C1"`.

````python
# Write your solution here
def draw_board(board):
    n = len(board)
    lines = []

    header = "  "
    for col in range(n):
        header += chr(ord('A') + col) + " "
    lines.append(header)

    for i in range(n):
        line = str(i + 1) + " "
        for j in range(n):
            if board[i][j] == 1:
                line += "X "
            elif board[i][j] == 2:
                line += "O "
            else:
                line += "  "
        lines.append(line)

    return "\n".join(lines)
'''
https://www.w3schools.com/python/ref_func_ord.asp
'''
````

*Exercise 7:* Write a function that takes a board, a player number, and a location string (as in Exercise 6), then uses your function from Exercise 5 to update the board.

````python
# Write your solution here
def place_move_from_location(board, player, location):
    col = ord(location[0].upper()) - ord('A')
    row = int(location[1]) - 1

    return place_move(board, player, row, col)
````

````python
# Test your solution here
board = create_board_square_matrix(3)

print(place_move_from_location(board, 1, "A1"))
print(place_move_from_location(board, 2, "B2"))
print(place_move_from_location(board, 1, "A1"))

print(draw_board(board))
````

*Exercise 8:* Write a function that is called with a board and player number, takes input from the player using Python's `input()`, and modifies the board using your function from Exercise 7.

Keep asking for input until the player enters a valid location that results in a valid move.


````python
# Write your solution here
def player_turn(board, player):
    while True:
        location = input(f"Player {player}, enter a location: ")

        try:
            empty_cell = place_move_from_location(board, player, location)
            if empty_cell:
                return
            else:
                print("That spot is already taken. Try again.")
        except:
            print("Invalid location. Try again.")
````

````python
# Test your solution here

board = create_board_square_matrix(3)

print(draw_board(board))
player_turn(board, 1)
print(draw_board(board))
````

````python
# Write your solution here
def play_tic_tac_toe(n):
    board = create_board_square_matrix(n)
    player = 1

    while True:

        print(draw_board(board))


        player_turn(board, player)


        result = check_winner(board)

        if result == 1:
            print(draw_board(board))
            print("Player 1 (X) wins!")
            break
        elif result == 2:
            print(draw_board(board))
            print("Player 2 (O) wins!")
            break
        elif result == 0:
            print(draw_board(board))
            print("The game is a draw.")
            break


        if player == 1:
            player = 2
        else:
            player = 1


````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
# Write your solution here
#Took help from chat gpt with the prompt, exhaustive search with max depth parameter for the above code and pasted the above code. it provided with def minimax and flow on how to code this program which I compiled.
def minimax(board, depth, is_maximizing, player, max_depth):
    result = check_winner(board)
    opponent = 1 if player == 2 else 2

    if result == player:
        return 10 - depth
    if result == opponent:
        return depth - 10
    if result == 0 or depth == max_depth:
        return 0

    n = len(board)

    if is_maximizing:
        best_score = -1000
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = player
                    score = minimax(board, depth + 1, False, player, max_depth)
                    board[i][j] = 0
                    best_score = max(best_score, score)
        return best_score
    else:
        best_score = 1000
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = opponent
                    score = minimax(board, depth + 1, True, player, max_depth)
                    board[i][j] = 0
                    best_score = min(best_score, score)
        return best_score


def computer_move(board, player, max_depth):
    n = len(board)
    best_score = -1000
    best_move = None

    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = player
                score = minimax(board, 0, False, player, max_depth)
                board[i][j] = 0
                if score > best_score:
                    best_score = score
                    best_move = (i, j)

    if best_move:
        board[best_move[0]][best_move[1]] = player


def play_tic_tac_toe_vs_computer(n, max_depth):
    board = create_board_square_matrix(n)
    player = 1

    while True:
        print(draw_board(board))

        if player == 1:
            player_turn(board, 1)
        else:
            computer_move(board, 2, max_depth)

        result = check_winner(board)

        if result == 1:
            print(draw_board(board))
            print("Player 1 (X) wins!")
            return
        elif result == 2:
            print(draw_board(board))
            print("Computer (O) wins!")
            return
        elif result == 0:
            print(draw_board(board))
            print("The game is a draw.")
            return

        player = 2 if player == 1 else 1
````
