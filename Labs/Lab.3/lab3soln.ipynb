````python
# Skeleton
def variance(Data):
    v = 0.0

    ### BEGIN SOLUTION

    if len(Data) == 0:
        return 0.0

    mu = mean(Data)

    v = sum((x - mu) ** 2 for x in Data) / len(Data)

    ### END SOLUTION

    return v
````

## Histogramming

Now we’ll turn raw numbers into binned counts so we can *see* the distribution.


**Exercise 3:**  
Implement `histogram(x, n_bins=10, x_min=None, x_max=None)` that returns:
- `hist`: a list of counts per bin
- `bin_edges`: a list of bin edge values (length `n_bins+1`)

Use evenly spaced bins between `x_min` and `x_max` (or infer them from the data if not provided).


````python
# Solution
def histogram(x, n_bins=10, x_min=None, x_max=None):
    ### BEGIN SOLUTION
    # if list is empty, return zero counts
    if len(x) == 0:
        return [0]*n_bins, [0.0]*(n_bins+1)

    # set lower and upper bounds
    lo = min(x) if x_min is None else x_min
    hi = max(x) if x_max is None else x_max

    # avoiding zero division if all values are equal
    if hi == lo:
        hi = lo + 1e-12

    # computing bin width and edges
    width = (hi - lo) / n_bins
    bin_edges = [lo + i * width for i in range(n_bins + 1)]
    hist = [0] * n_bins

    # counting how many values fall in each bin
    for val in x:
        idx = int((val - lo) / width)

        # making sure max value goes in last bin
        if idx == n_bins:
            idx = n_bins - 1

        if 0 <= idx < n_bins:
            hist[idx] += 1
    ### END SOLUTION

    return hist, bin_edges
````

````python
# Test your solution here
h, edges = histogram(data, n_bins=20)
print("Counts (first 10 bins):", h[:10])
print("Bin edges (first 6):", [round(e, 2) for e in edges[:6]])
````

**Exercise 4:**  
Write `draw_histogram(...)` that prints a simple text bar chart.  
Example idea (your spacing can differ):

```
[0.0, 0.5)  ####
[0.5, 1.0)  ##########
```


````python
# Solution
def draw_histogram(x, n_bins, x_min=None, x_max=None, character="#", max_character_per_line=20):
    ### BEGIN SOLUTION
    # to get histogram data first
    hist, bin_edges = histogram(x, n_bins=n_bins, x_min=x_min, x_max=x_max)

    # finding the largest count for scaling
    max_count = max(hist) if len(hist) else 1

    for i, count in enumerate(hist):
        left = bin_edges[i]
        right = bin_edges[i+1]

        # scaling bar length relative to largest bin
        bar_len = 0 if max_count == 0 else int(round((count / max_count) * max_character_per_line))

        bar = character * bar_len
        print(f"[{left:.3f}, {right:.3f}) {bar} ({count})")
    ### END SOLUTION

    return hist, bin_edges
````

## Functional Programming

Here we practice passing functions *into* other functions.


````python
def where(mylist, myfunc):
    out = []

    ### BEGIN SOLUTION
    # creating a loop through list with index
    for i, val in enumerate(mylist):

        # to check condition
        if myfunc(val):
            out.append(i)
    ### END SOLUTION

    return out
````

````python
# Test your solution here
demo = [0.1, 0.7, 0.2, 0.9, 0.51]
idx = where(demo, lambda t: t > 0.5)
print("Indices where value > 0.5:", idx)
print("Values:", [demo[i] for i in idx])
````

**Exercise 6:**  
The `in_range(mymin, mymax)` pattern returns a *function* that checks whether a value is in that range.

Create similar “function builders” for:
- even
- odd
- greater_than
- less_than
- equal_to
- divisible_by


````python
def in_range(mymin, mymax):
    def test_range(x):
        return (x < mymax) and (x >= mymin)

    return test_range

# Examples:
F1 = in_range(0, 10)
F2 = in_range(10, 20)

# Test of in_range
print(F1(0), F1(1), F1(10), F1(9.999))
print(F2(9.999), F2(10), F2(19.9), F2(20))
````

````python
### BEGIN SOLUTION
def is_even():
    return lambda x: (x % 2) == 0

def is_odd():
    return lambda x: (x % 2) != 0

def greater_than(threshold):
    return lambda x: x > threshold

def less_than(threshold):
    return lambda x: x < threshold

def equal_to(target):
    return lambda x: x == target

def divisible_by(k):
    return lambda x: (x % k) == 0
### END SOLUTION
````

````python
# Test your solution
nums = list(range(-5, 16))

print("Even examples:", [n for n in nums if is_even()(n)][:8])
print("Odd examples:", [n for n in nums if is_odd()(n)][:8])
print("> 7:", [n for n in nums if greater_than(7)(n)])
print("< 0:", [n for n in nums if less_than(0)(n)])
print("== 3:", [n for n in nums if equal_to(3)(n)])
print("Divisible by 4:", [n for n in nums if divisible_by(4)(n)])
````

## Monte Carlo

We’ll generate random samples that follow a target function using rejection sampling.


````python
import random

def generate_function(func, x_min, x_max, N=1000):
    out = list()
    ### BEGIN SOLUTION

    # to use rejection sampling

    # we estimate max value of func on the interval
    probes = 2000
    xs = [random.uniform(x_min, x_max) for _ in range(probes)]
    f_max = max(func(x) for x in xs)

    if f_max <= 0:
        raise ValueError("func must be positive on the sampling interval.")

    # sampling until we collect N points
    while len(out) < int(N):
        x = random.uniform(x_min, x_max)
        y = random.uniform(0, f_max)

        # accept x if it falls under the curve
        if y <= func(x):
            out.append(x)

    ### END SOLUTION

    return out
````

Use your generator to draw 1000 samples from a Normal distribution created by `gaussian(mean, sigma)`.  
Check that the sample mean/variance are reasonably close to the requested values, and histogram the samples.


````python
import math

def gaussian(mean, sigma):
    # defining normal distribution with given mean and standard deviation
    def f(x):
        return math.exp(-((x - mean) ** 2) / (2 * sigma ** 2)) / (sigma * math.sqrt(2 * math.pi))
    return f


# Example Instantiation
g1 = gaussian(0, 1)
g2 = gaussian(10, 3)
````

**Exercise 9:**  
Combine `generate_function`, `where`, and `in_range` to build an `integrate` function that estimates probability mass in an interval.

Use it to show that roughly **68%** of a Normal distribution lies within **one standard deviation** of the mean.


````python
def integrate(func, x_min, x_max, n_points=5000):
    # first to estimate probability mass using Monte Carlo sampling

    # we expand the sampling range to include tails
    span = x_max - x_min
    gen_min = x_min - 5 * span
    gen_max = x_max + 5 * span

    # then generate samples from the function
    samples = generate_function(func, gen_min, gen_max, N=n_points)

    # counting how many fall inside the target interval
    idx = where(samples, in_range(x_min, x_max))

    integral = len(idx) / len(samples) if len(samples) else 0.0

    return integral
````

````python
# Exercise 9: probability within 1 sigma (should be close to ~0.68)
mu, sig = 0.0, 1.0
g = gaussian(mu, sig)

p = integrate(g, mu - sig, mu + sig, n_points=8000)
print("Estimated P(|X-mu| <= sigma):", round(p, 4))
````
